```python
Arx=[] # The possible states are represented by a list
Telos=[1,2,3,4,5,6,7,8,0] #The goal state
print('To μηδέν αναπαριστά το κενό') #Zero represents the space
Arx.append(int(input('πάνω αριστερά'))) # The user gives the initial state, the first element corresponds to the upper left piece
Arx.append(int(input('πάνω'))) #the second element corresponds to the upper piece
Arx.append(int(input('πάνω δεξια')))#the third element corresponds to the upper right piece
Arx.append(int(input('αριστερά'))) #the fourth element corresponds to the left piece...
Arx.append(int(input('κέντρο')))
Arx.append(int(input('δεξιά')))
Arx.append(int(input('κάτω αριστερά')))
Arx.append(int(input('κάτω')))
Arx.append(int(input('κάτω δεξιά')))

print(Arx) #Prints start state
mesi=Arx.copy() #Makes a copy of the initial state

def up(mesi):  # The action 'move the space up' is emulated by a function
    mesi_u=mesi.copy() 
    if 0 in mesi[3:9]: # If the space is not in the upper line, it will be executed 
        thesikeno=mesi.index(0) #finds where the zero is
        mesi_u[thesikeno]=mesi[thesikeno-3] # changes the value of the space brick with the above it
        mesi_u[thesikeno-3]=0 #and the one above with 0
    else:#If the space is in the upper line, it won't be executed 
        pass
    return mesi_u
def down(mesi): # The action down
    mesi_d=mesi.copy()
    if 0 in mesi[0:6]:
        thesikeno=mesi.index(0)
        mesi_d[thesikeno]=mesi[thesikeno+3]
        mesi_d[thesikeno+3]=0
    else:
        pass
    return mesi_d
def right(mesi): # The action right
    mesi_r=mesi.copy()
    if mesi[2]!=0 and mesi[5]!=0 and mesi[8]!=0:
        thesikeno=mesi.index(0)
        mesi_r[thesikeno]=mesi[thesikeno+1]
        mesi_r[thesikeno+1]=0
    else:
        pass
    return mesi_r
def left(mesi): # The action left
    mesi_l=mesi.copy()
    if mesi[0]!=0 and mesi[3]!=0 and mesi[6]!=0:
        thesikeno=mesi.index(0)
        mesi_l[thesikeno]=mesi[thesikeno-1]
        mesi_l[thesikeno-1]=0
    else:
        pass
    return mesi_l
        
def ektim(mesi): #An admissible heuristic (Manhatan distance)
    ektim=0
    for i in range(1,9): #For every piece
        ektim+=abs(mesi.index(i)//3-(i-1)//3)+abs(mesi.index(i)%3-(i-1)%3) # The distance in colums and rows of the ideal and current position
    return ektim

diadromi={0:[ektim(mesi),mesi]} #The search tree search. A key, evaluation, the action sequence(cannot be seen), and the state
fylla=[[ektim(mesi)+0,mesi,0,0]] #Leave nodes. Evaluation plus steps to reach the node(A*), the state, the key, and the parent node
palies=[mesi] #Visited states

def elax(fylla): # Finds the minimum evaluation
    elaxi=[]
    for i in range(len(fylla)):
        elaxi.append(fylla[i][0])
    return min(elaxi)

def epilogi(fylla): #chooses the firt node of the leave nodes with evaluation equal to the minimum
    i=0
    while i<len(fylla):
        if fylla[i][0]==elax(fylla):
            break
        i+=1
    return i
            

def ekser(fylla,i): #expands the optimal node
    kratoymeno=1 
    a=diadromi[fylla[i][2]].copy()
    b=diadromi[fylla[i][2]].copy()
    c=diadromi[fylla[i][2]].copy()
    d=diadromi[fylla[i][2]].copy()
    if up(fylla[i][1])!=fylla[i][1] and fylla[i][3]!=1 and up(fylla[i][1]) not in palies:#if acion up has efect and doesn't return to state that is explored
        a[-1]=up(fylla[i][1]) 
        a[0]=ektim(up(fylla[i][1]))
        a.insert(-1,"up")
        diadromi[fylla[i][2]]=a # Add the new state to the memory
        palies.append(up(fylla[i][1]))
        neaolekt=fylla[i][0]+ektim(up(fylla[i][1]))-ektim(fylla[i][1])+1 #changes the new total evaluation of the new state
        fylla.append([neaolekt,up(fylla[i][1]),fylla[i][2],2])
        kratoymeno=0
    else:
        pass
    if down(fylla[i][1])!=fylla[i][1] and fylla[i][3]!=2 and down(fylla[i][1]) not in palies:
        b[-1]=down(fylla[i][1])
        b[0]=ektim(down(fylla[i][1]))
        b.insert(-1,"down")
        palies.append(down(fylla[i][1]))
        neaolekt=fylla[i][0]+ektim(down(fylla[i][1]))-ektim(fylla[i][1])+1
        if kratoymeno==0:
            fylla.append([neaolekt,down(fylla[i][1]),len(diadromi),1])
            diadromi[len(diadromi)]=b
        else:
            diadromi[fylla[i][2]]=b
            fylla.append([neaolekt,down(fylla[i][1]),fylla[i][2],1])
            kratoymeno=0
    else:
        pass
    if right(fylla[i][1])!=fylla[i][1] and fylla[i][3]!=3 and right(fylla[i][1]) not in palies:
        c[-1]=right(fylla[i][1])
        c[0]=ektim(right(fylla[i][1]))
        c.insert(-1,"right")
        palies.append(right(fylla[i][1]))
        neaolekt=fylla[i][0]+ektim(right(fylla[i][1]))-ektim(fylla[i][1])+1
        if kratoymeno==0:
            fylla.append([neaolekt,right(fylla[i][1]),len(diadromi),4])
            diadromi[len(diadromi)]=c
        else:
            fylla.append([neaolekt,right(fylla[i][1]),fylla[i][2],4])
            diadromi[fylla[i][2]]=c
            kratoymeno=0
    else:
        pass
    if left(fylla[i][1])!=fylla[i][1] and fylla[i][3]!=4 and left(fylla[i][1]) not in palies:
        d[-1]=left(fylla[i][1])
        d[0]=ektim(left(fylla[i][1]))
        d.insert(-1,"left")
        palies.append(left(fylla[i][1]))
        neaolekt=fylla[i][0]+ektim(left(fylla[i][1]))-ektim(fylla[i][1])+1
        if kratoymeno==0:
            fylla.append([neaolekt,left(fylla[i][1]),len(diadromi),3])
            diadromi[len(diadromi)]=d
        else:
            fylla.append([neaolekt,left(fylla[i][1]),fylla[i][2],3])
            diadromi[fylla[i][2]]=d 
    else:
        pass
    del fylla[i]
    return fylla

i=0
while len(fylla)>0 and Telos not in palies[-4:]: #While Goal state not found (and the tree is not completly searched)
    ekser(fylla,epilogi(fylla)) #expand the optimal leave node
    i+=1
    if i%100==0: #Print once in a while how many leave are
        print(len(fylla))
    

def teleiwsame(fylla): # find the goal state in the leaves
    j=0
    while j<len(fylla):
        if fylla[j][1]==Telos:
            break
        else:
            pass
        j+=1
    return fylla[j][2] #And give the the key




print(diadromi[teleiwsame(fylla)]) #use the Key to print the sequence
```
